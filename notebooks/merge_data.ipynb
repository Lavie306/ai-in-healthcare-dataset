```python
import pandas as pd
import json
```

```python
df1 = pd.read_csv("../data_clean/medical_dataset.csv")
df2 = pd.read_csv("../data_clean/medical_data_combined.csv")
df3 = pd.read_csv("../data_clean/ViHealthQA_combined_cleaned.csv")

print("medical_dataset:", len(df1))
print("medical_data_combined:", len(df2))
print("ViHealthQA_combined:", len(df3))
```

```python
df = pd.concat([df1, df2, df3], ignore_index=True)

print("Total QA after merge:", len(df))
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()

print("After removing null:", len(df))
```

```python
df_clean = df.drop_duplicates(subset=["question", "answer"])

print("After removing exact QA duplicates:", len(df_clean))

print("Removed rows:", len(df) - len(df_clean))
```

```python
duplicates = df_clean.duplicated(subset=["question", "answer"]).sum()

print("Remaining duplicate QA:", duplicates)
```

```python
print("Total QA:", len(df_clean))
print("Unique questions:", df_clean["question"].nunique())
print("Unique answers:", df_clean["answer"].nunique())
```

```python
df_clean.to_csv("../data_processed/final_medical_qa_dataset.csv", index=False)
```

```python
with open("../data_processed/final_medical_qa_dataset.json", "w", encoding="utf8") as f:
    json.dump(df_clean.to_dict(orient="records"), f, ensure_ascii=False, indent=2)

print("Saved final dataset!")
```

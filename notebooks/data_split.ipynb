```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../data_processed/final_medical_qa_dataset.csv")

print("Total samples:", len(df))
df.head()
```

```python
train_df, temp_df = train_test_split(
    df,
    test_size=0.2,
    random_state=42
)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    random_state=42
)
```

80% train
10% validation
10% test

```python
print("Train size:", len(train_df))
print("Validation size:", len(val_df))
print("Test size:", len(test_df))
```

```python
train_df.to_csv("../data_processed/train.csv", index=False)
val_df.to_csv("../data_processed/val.csv", index=False)
test_df.to_csv("../data_processed/test.csv", index=False)
```

```python
from src.load_data import load_csv
from src.preprocessing import clean_text
from src.preprocessing import clean_dataframe
import pandas as pd
```

```python
df1 = load_csv('../data_raw/ViHealthQA/train.csv')
df2 = load_csv('../data_raw/ViHealthQA/test.csv')
df3 = load_csv('../data_raw/ViHealthQA/val.csv')
```

```python
# Gộp 3 tập train, test, val thành một bộ dữ liệu lớn
df_all = pd.concat([df1, df2, df3], ignore_index=True)

# Loại bỏ cột 'id' và 'link' nếu không cần thiết cho việc huấn luyện mô hình
df_all = df_all[['question', 'answer']]

# Sử dụng hàm clean_dataframe (giả định hàm này xử lý null và áp dụng clean_text)
# Nếu chưa có, bạn có thể áp dụng trực tiếp:
df_all = df_all.dropna()
df_all['question'] = df_all['question'].apply(clean_text)
df_all['answer'] = df_all['answer'].apply(clean_text)

# Loại bỏ trùng lặp
df_all = df_all.drop_duplicates()
```

```python
# Lưu CSV (Sử dụng utf-8-sig để hiển thị tốt trên Excel)
df_all.to_csv('../data_clean/ViHealthQA_combined_cleaned.csv', index=False, encoding='utf-8-sig')

# Lưu JSON (orient='records' để tạo danh sách các object)
df_all.to_json('../data_clean/ViHealthQA_combined_cleaned.json', orient='records', force_ascii=False, indent=4)

print(f"Tổng số dòng sau khi gộp và làm sạch: {len(df_all)}")
```
